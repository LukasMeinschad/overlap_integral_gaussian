```python
import numpy as np

# Compute Eigenvalues and Eigenvectors

matrix = np.array([[1,2], [3,4]])

print("Matrix:\n", matrix)

eigenvalues, eigenvectors = np.linalg.eig(matrix)

print("Eigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)


```

```python
# Generatr 3 Vectors and compute length

v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])
v3 = np.array([7, 8, 9])

norm1 = np.linalg.norm(v1)
norm2 = np.linalg.norm(v2)
norm3 = np.linalg.norm(v3)

print("Norm of v1:", norm1)
print("Norm of v2:", norm2)
print("Norm of v3:", norm3)

# Calculate with different p norms
p1 = np.linalg.norm(v1, 1)
p2 = np.linalg.norm(v2, 2)
p3 = np.linalg.norm(v3, np.inf)

print("1-Norm of v1:", p1)
print("2-Norm of v2:", p2)
print("Infinity Norm of v3:", p3)
```

```python
import matplotlib.pyplot as plt

def quadratic_function(x):
    return x**2 + 4

x = np.linspace(-10, 10, 100)
print(type(x))
y = quadratic_function(x)

plt.plot(x, y)
plt.title('Quadratic Function: y = x^2 + 4')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
